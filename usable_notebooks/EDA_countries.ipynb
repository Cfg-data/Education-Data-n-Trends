```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# URLs for the CSV files
url_df = "https://raw.githubusercontent.com/Cfg-data/final-project/refs/heads/master/data/merged/final/SYB67_245_202411final_data.csv"
url_df1 = "https://raw.githubusercontent.com/Cfg-data/final-project/refs/heads/master/data/merged/final/SYB67_309_202411final_data.csv"
url_df2 = "https://raw.githubusercontent.com/Cfg-data/final-project/refs/heads/master/data/merged/final/SYB67_319_202411final_data.csv"
url_df3 = "https://raw.githubusercontent.com/Cfg-data/final-project/refs/heads/master/data/merged/final/SYB67_323_202411final_data.csv"

# Load the CSV files into pandas DataFrames
df = pd.read_csv(url_df)
df1 = pd.read_csv(url_df1)
df2 = pd.read_csv(url_df2)
df3 = pd.read_csv(url_df3)
```

```python
# Merge all DataFrames on 'year' and 'country' (if applicable)
merged_df = pd.merge(df, df1, on=['Year', 'Region/Country/Area'], suffixes=('_df', '_df1'))
merged_df = pd.merge(merged_df, df2, on=['Year', 'Region/Country/Area'], suffixes=('', '_df2'))
merged_df = pd.merge(merged_df, df3, on=['Year', 'Region/Country/Area'], suffixes=('', '_df3'))

output_path = 'full_country_data.csv'

# Save the merged DataFrame to a CSV file
merged_df.to_csv(output_path, index=False)

# Print confirmation message
print(f"Merged DataFrame has been saved to: {output_path}")
```

```python
# Compute the correlation matrix for numerical columns
correlation_matrix = merged_df.corr()

# Set up the plot
plt.figure(figsize=(18, 16))

# Create a heatmap of the correlation matrix with a dark cyan color scheme
sns.heatmap(correlation_matrix, cmap='winter', annot=True, fmt=".2f", linewidths=0.5, cbar_kws={"shrink": 0.75}, square=True)

# Customize the title and labels
plt.title('Correlation Matrix by Year for chosen Country Codes', fontsize=16)

# Define the output path for saving the figure
output_folder = r"C:\Users\Cecilia\Downloads\ironhack\coursework\Final_Project\usable_notebooks\correlations"
output_path = f"{output_folder}\\correlation_matrix.png"

# Save the plot as a PNG file and trim the white space
plt.savefig(output_path, bbox_inches='tight', pad_inches=0.1)

# Show the plot
plt.show()

# Print confirmation message
print(f"Correlation matrix plot saved to: {output_path}")
```

```python
# Exclude columns 'Year' and 'Region/Country/Area' for statistical analysis
columns_to_analyze = [col for col in merged_df.columns if col not in ['Year', 'Region/Country/Area']]

# Group the data by 'Year' and 'Region/Country/Area' and compute summary statistics
grouped = merged_df.groupby(['Year', 'Region/Country/Area'])[columns_to_analyze].describe()

# Reset the index to make the table easier to read
grouped = grouped.reset_index()

# Display the statistical summary table
grouped.head()
```

```python
# Define the output file name in the working directory
output_path = 'statistical_analysis_countries.csv'

# Save the statistical analysis table to a CSV file
grouped.to_csv(output_path, index=False)

# Print confirmation message
print(f"Statistical analysis table has been saved to: {output_path}")
```
