```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/Cfg-data/final-project/refs/heads/master/data/cleaned/cleaned_SYB67_245_202411.csv"

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(url)

# Display the first few rows of the dataframe
df.head()
```

```python
# List of region/country/area codes to filter
codes = [8, 20, 40, 36, 53, 56, 70, 100, 112, 124, 156, 191, 196, 203, 208, 233, 
         246, 250, 268, 276, 292, 308, 344, 348, 352, 356, 372, 380, 392, 410, 
         428, 438, 440, 442, 446, 470, 484, 492, 498, 499, 528, 554, 578, 616, 
         620, 642, 643, 688, 703, 705, 724, 752, 756, 804, 826, 840, 674]

# Assuming the column for region/country/area codes is named 'RegionCode' (change as needed)
filtered_df = df[df['Region/Country/Area'].isin(codes)]

# Reset the index
filtered_df.reset_index(drop=True, inplace=True)

# Display the filtered dataframe
filtered_df.head()
```

```python
# Define the file path for saving the CSV file
file_path = r'C:\Users\Cecilia\Downloads\ironhack\coursework\Final_Project\data\filtered\filtered_countries_SYB67_245_202411.csv'

# Save the filtered DataFrame to a CSV file
filtered_df.to_csv(file_path, index=False)

print(f"Filtered data saved to {file_path}")
```
