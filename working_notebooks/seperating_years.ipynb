```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/Cfg-data/final-project/refs/heads/master/data/filtered/filtered_countries_SYB67_245_202411.csv"

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(url)

# Display the first few rows of the dataframe
df.head()
```

```python
# List of country codes to check
country_codes = [8, 20, 40, 36, 53, 56, 70, 100, 112, 124, 156, 191, 196, 203, 208, 233, 246, 250, 268, 
                 276, 292, 308, 344, 348, 352, 356, 372, 380, 392, 410, 428, 438, 440, 442, 446, 470, 484, 
                 492, 498, 499, 528, 554, 578, 616, 620, 642, 643, 688, 703, 705, 724, 752, 756, 804, 826, 
                 840, 674]

# Get the unique years in the dataset
all_years = df['Year'].unique()

# Filter the DataFrame for the country codes
df_filtered = df[df['Region/Country/Area'].isin(country_codes)]

# Initialize a dictionary to store missing years for each country code
missing_years_dict = {}

# For each country code, check which years are missing
for country_code in country_codes:
    country_data = df_filtered[df_filtered['Region/Country/Area'] == country_code]
    country_years = country_data['Year'].unique()
    missing_years = set(all_years) - set(country_years)
    
    if missing_years:  # Only add to the dictionary if there are missing years
        missing_years_dict[country_code] = sorted(list(missing_years))

# Print the missing years for each country code
for country_code, missing_years in missing_years_dict.items():
    print(f"Country Code {country_code} is missing years: {missing_years}")
```
