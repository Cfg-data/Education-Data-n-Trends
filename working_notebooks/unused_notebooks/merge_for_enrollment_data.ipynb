```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/Cfg-data/final-project/refs/heads/master/data/raw/enrollment_data/UNdata_Export_20241213_133629943.csv"

# Load the CSV file into a pandas DataFrame
df = pd.read_csv(url)

# Display the first few rows of the dataframe
df.head()
```

```python
url = "https://raw.githubusercontent.com/Cfg-data/final-project/refs/heads/master/data/raw/country_codes_file.csv"

# Load the CSV file into a pandas DataFrame
df1 = pd.read_csv(url)

# Display the first few rows of the dataframe
df1.head()
```

```python
# Merge the 'Country/Area Code' from df1 into df by matching 'Reference Area' in df to 'Country/Area Name' in df1
merged_df = pd.merge(df, df1[['Country/Area Name', 'Country/Area Code']], left_on='Reference Area', right_on='Country/Area Name', how='left')

# Display the merged DataFrame
print("\nMerged DataFrame:")
merged_df.head()
```

```python
# Print unique values for all columns
for column in merged_df.columns:
    print(f"Unique values in column '{column}':")
    print(merged_df[column].unique())
    print() 
```

```python
# List of regions/areas to filter
areas_to_extract = [
    'MDG_Caribbean', 'MDG_Caucasus and Central Asia', 'MDG_Developed regions',
    'MDG_Developing regions', 'MDG_Eastern Asia', 'MDG_Landlocked developing countries',
    'MDG_Latin America', 'MDG_Latin America and the Caribbean', 'MDG_Least developed countries',
    'MDG_Northern Africa', 'MDG_Oceania', 'MDG_Small island developing States',
    'MDG_South-eastern Asia', 'MDG_Southern Asia', 'MDG_Southern Asia excluding India',
    'MDG_Sub-Saharan Africa', 'MDG_Western Asia'
]

# Filter the merged DataFrame to only include rows where 'Reference Area' is in the list of areas
filtered_df = merged_df[merged_df['Reference Area'].isin(areas_to_extract)]

# Display the filtered DataFrame
print(filtered_df)

# Print filtered data to a CSV file
filtered_df.to_csv("filtered_areas_data.csv", index=False)

print("Filtered data has been saved to 'filtered_areas_data.csv'")
```

```python
# Drop the rows that were filtered (rows where 'Reference Area' is in areas_to_extract)
remaining_df = merged_df[~merged_df['Reference Area'].isin(areas_to_extract)]

remaining_df.head()
```

```python
# Get unique values in 'Reference Area' from remaining_df
reference_area_unique_remaining = remaining_df['Reference Area'].unique()

# Get unique values in 'Country/Area Name' from df1
country_area_name_unique = df1['Country/Area Name'].unique()

# Find differences: Values in 'Reference Area' not in 'Country/Area Name'
reference_area_not_in_country = set(reference_area_unique_remaining) - set(country_area_name_unique)

# Find values in 'Country/Area Name' not in 'Reference Area'
country_area_not_in_reference = set(country_area_name_unique) - set(reference_area_unique_remaining)

# Print the differences
print("Values in 'Reference Area' (from remaining_df) not in 'Country/Area Name':")
print(reference_area_not_in_country)

print("\nValues in 'Country/Area Name' (from df1) not in 'Reference Area' (from remaining_df):")
print(country_area_not_in_reference)
```
