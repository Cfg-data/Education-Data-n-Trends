```python
url = "https://raw.githubusercontent.com/Cfg-data/final-project/refs/heads/master/data/cleaned/SYB67_323_202411_cleaned.csv"

# Load the CSV file into a pandas DataFrame
df1 = pd.read_csv(url)

# Display the first few rows of the dataframe
df1.head()
```

```python
# Keep the 'Region/Country/Area' column and make it the first column
df1 = df1[['Region/Country/Area'] + [col for col in df1.columns if col != 'Region/Country/Area']]
# Keep the 'Region/Country/Area' column and make it the first column in df2
df = df[['Region/Country/Area'] + [col for col in df.columns if col != 'Region/Country/Area']]
# Merge df1 and df2 on the 'Region/Country/Area' column
merged_df = pd.merge(df, df1[['Region/Country/Area']], on='Region/Country/Area', how='left')

# Ensure 'Region/Country/Area' is the first column in the merged DataFrame
merged_df = merged_df[['Region/Country/Area'] + [col for col in merged_df.columns if col != 'Region/Country/Area']]

# Display the merged DataFrame
print("\nMerged DataFrame:")
print(merged_df.head())
```
